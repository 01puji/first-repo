## **[3] Data 분석**

- Data 구조 분석(행x열 크기 출력)

```python
df.shape                                     # Data 구조 분석(행x열 크기 출력)
```

- 열의 제목 분석

```python
df.columns                                    # 모든 열의 제목 나열
```

- Data의 전체적인 정보 분석

```python
df.info()                                      # 열별 데이터 분석
```

- 특정 열에 대한 분석

```python
df['LotShape'].value_counts()                    # 'LotShape' 열 데이터 분석
```

- 특정 위치의 데이터 뽑아서 출력

```python
df.loc[5,'LotFrontage']                          # 'LotFrontage'열의 6번째 데이터 출력
```

- 특정 구간의 데이터 뽑아서 출력

```python
df.loc[2:5,'LotFrontage']                        # 'LotFrontage'열의 3,4,5,6번째 데이터 출력
```

- 특정 영역의 데이터 뽑아서 출력

```python
df.loc[2:5,['LotFrontage','LotShape']]          # 'LotFrontage'열과 'LotShape'열의 3,4,5,6번째 데이터 출력
```

- 특정 조건을 만족하는 데이터 뽑아서 출력

```python
cond = (df['LotFrontage'] >= 100)               # 'LotFrontage'가 100 이상인 데이터 출력
df.loc[cond]
```
